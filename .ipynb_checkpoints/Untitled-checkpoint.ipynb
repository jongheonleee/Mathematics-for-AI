```python
from sympy import Limit, S, Symbol

# 변수 x 생성, 0으로 한없이 가까워질때
# 1/x 극한값 구함
x = Symbol('x')
Limit(1/x, x, S.Infinity).doit()
```

```python
# 우극한 값 구하기
Limit(1/x, x, 0).doit()
```

```python
# 좌극한 값 구하기
Limit(1/x, x, 0, dir='-').doit()
```
