```python
# 방정식과 부등식

'''
-SymPy

> 파이썬에서 기호 수학을 위한 라이브러리

  1) 속도, 시각화 등에 필요한 확장 기능도 포함
  
  2) 대수 문제를 기호 수학으로 풀 수 있음
  
'''
from sympy import Symbol, solve

x = Symbol('x')

equ = 2 * x - 6

# 방정식을 풀려면 solve() 함수를 사용하면됨
solve(equ)
```

```python
# solve()함수 활용
# 일단 equ = '방정식', 좌변에 변수 할당(0)이고 우변에는 방정식 모두 다 이항해서 넣어주기

k = Symbol('k')
equ1 = k-2-4
solve(equ1)
```

```python
equ2 = k -16
solve(equ2)
```

```python
# 방정식 vs 항등식: 방정식은 등호(=)를 기준으로 미지수에따라 참/거짓으로 나뉨, 항등식은 좌변, 우변이 항상 참인 등식
```

```python
# 연립방정식: 미지수가 여러 개 포함된 방정식을 묶어 놓은 것 (미지수가 2개면 최소 식도 2개, 3개면 최소 식도 3개)
# 소거해주는게 포인트임

# 파이썬에서 연립방정식 구하기

x = Symbol('x')
y = Symbol('y')

e1 = 3 * x + y - 2
e2 = x - 2 * y - 3

solve((e1, e2), dict=True)
# 해를 딕셔너리 형태로 반환해줌
```
