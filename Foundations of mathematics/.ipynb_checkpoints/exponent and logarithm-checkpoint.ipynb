```python
# exponent and logarithm
'''
오차역전파에 사용하는 수학
'''

import math

# 2의 5 제곱
pow = math.pow(2, 5)
print('pow 결과:', pow)
```

```python
# 4의 제곱근
sqrt = math.sqrt(4)
print('sqrt 결과:', sqrt)
```

```python
# e의 2 제곱
exp = math.exp(2)
print('exp 결과:', exp)
```

```python
# log값을 얻으려고 math 라이브러리의 log() 함수 활용
# 밑이 4이고, 진수가2인 로그
# log4 2
math.log(2, 4)
```

```python
# 밑이 2이고, 진수가 4인 로그
math.log(4, 2)
```
