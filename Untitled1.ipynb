```python
from sympy import symbols

def avg(a, b):
    m, n = max(a, b), min(a, b)
    
    x = symbols('x')
    
    fx = 2 * x ** 2 + 4 * x + 7
    fb = fx.subs(x, m)
    fa = fx.subs(x, n)
    
    res = (fb-fa) / b-a
    
    return res

print(avg(0, 2))
```

```python
from sympy import Derivative, symbols

x = symbols('x')
fx = 2 * x ** 2 + 4 * x + 7

fprime = Derivative(fx, x).doit()
n = fprime.subs({x:3})

print('fx에서 x = 3에서의 미분계수는', n)
```

```python
from sympy import Derivative, symbols

# 평균변화율을 구할 수 있는 함수를 정의함
x = symbols('x') # x를 기호변수화
fx = 2 * x ** 2 + 4 * x + 7


# Derivative() 객체 생성
# subs() 메서드를 사용하여 x=3에서의 미분계수 f'(3)을 구함
fprime = Derivative(fx, x).doit() # x에 대해 미분
n = fprime.subs({x: 3})

print('fx에서 x=3에서의 순간변화율(미분계수)', n, "입니다")



```

```python
import sympy as sym

# x 변수를 선언하고 diff() 함수를 사용하여 미분을 계산
# x = sym.Symbol('x')
x = symbols('x')
a = sym.diff((2 * x ** 3 + 3 * x ** 2 + x + 1), x)

print(a)
```
