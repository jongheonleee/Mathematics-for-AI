```python
from sympy import symbols

def avg(a, b):
    m, n = max(a, b), min(a, b)
    
    x = symbols('x')
    
    fx = 2 * x ** 2 + 4 * x + 7
    fb = fx.subs(x, m)
    fa = fx.subs(x, n)
    
    res = (fb-fa) / b-a
    
    return res

print(avg(0, 2))
```

```python
from sympy import Derivative, symbols

x = symbols('x')
fx = 2 * x ** 2 + 4 * x + 7

fprime = Derivative(fx, x).doit()
n = fprime.subs({x:3})

print('fx에서 x = 3에서의 미분계수는', n)
```
